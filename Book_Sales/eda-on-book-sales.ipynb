```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
# Loading the dataset
df = pd.read_csv('C:\\Users\\user\\OneDrive\\Documents\\Data_Scientist\\Practice\\EDA_Practice\\Book_Sales\\Books_Data_Clean.csv')
```

```python
df.head()
```

```python
# Numerical summary of numerical column
df.describe()
```

```python
# Fixing the negative value of publishing year as -560.000000
df = df[df["Publishing Year"] > 1900]
```

```python
df.describe()
```

```python
# Checking na values
df.isna().sum()
```

```python
# Dropping the book name rows which has no book name
df.dropna(subset = "Book Name",inplace = True)
```

```python
df.isna().sum()
```

```python
# Checking for dupicated values
df.duplicated().sum()
```

```python
# Checking unique
df.nunique()
```

```python
# Creating histogram of publishing year
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of publishing year")
plt.show()
```

```python
# ðŸ‘† From the above distribution we can observe that in our dataset we mostly have new books
```

```python
# Distribution of genres
df['genre'].value_counts().plot(kind = 'bar')
plt.xlabel("Genre")
plt.ylabel("Number of books")
plt.title("Number of books in each genre")
plt.show()
```

```python
# Calculating the avg rating 
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending = False)
```

```python
# Creating box plot of book rating count of each genre
sns.boxplot(x = 'genre',y = 'Book_ratings_count',data = df)
plt.xlabel('Genre')
plt.ylabel('Book Rating Count')
plt.title("Box Plot of book ratings count for each genre")
plt.show()
```

```python
# Scatter plot of sale price and unit sold
plt.scatter(df['sale price'],df['units sold'])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot Of Sale Price Vs Units Sold")
plt.show()
```

```python
# ðŸ‘† When the sale price drops the unit sold increases
```

```python
# Value counts of the langauge code
df['language_code'].value_counts().sort_values(ascending = True)
```

```python
# Creating pichart of language distribution of book
language_counts = df['language_code'].value_counts()
plt.pie(language_counts,labels = language_counts.index,startangle = 90,autopct = '&1.1f%%')
plt.title("Langauge distribution of book")
plt.show()
```

```python
# Grouping the data by publisher and calculate the total revenue for each publisher
df.groupby("Publisher ")['publisher revenue'].sum().sort_values(ascending = False)
```

```python
# Grouping the data by author rating and finding the avg of book ratings count for each author ratings.
df.groupby('Author_Rating')['Book_ratings_count'].mean().sort_values(ascending = False).max()
```

```python
# Group by language code and count the number of books in each lang
df.groupby('language_code').size().sort_values(ascending = False)
```

```python
# Group the data by author rating and find the max book ratings count for each author rating
df.groupby('Author_Rating')['Book_ratings_count'].max()
```

```python
# Scatter plot for book avg rating and book rating count
plt.scatter(df['Book_average_rating'],df['Book_ratings_count'])
plt.xlabel("Book Average Rating")
plt.ylabel("Book Rarting Count")
plt.title("Scatter plot of book avg rating vs book ratings count")
plt.show()
```

```python
# Creating bar chart for total grosss sales for each author
total_gross_sales_by_author = df.groupby('Author')['gross sales'].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending = False).head(10).plot(kind = 'bar')
plt.xlabel('Author')
plt.ylabel('Total Gross Sales')
plt.title('Total gross sales for each author')
plt.show()
```

```python
# Box plot for each unit sold for each author rating 
sns.boxplot(x="Author_Rating", y="units sold", data = df)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Box Plot of Units Sold for Each Author Rating")
plt.show()
```

```python
# Grouping the data by publishing year and plotting the line chart of total units
df.groupby('Publishing Year')['units sold'].sum().plot(kind = 'line',marker = 'o')
plt.xlabel("Publishing Year")
plt.ylabel("Total units sold")
plt.title("Total units sold over the years")
plt.show()
```
